This project focuses on fitting a cosine curve to noisy data using curve fitting techniques. The objective is to model the noisy data using a cosine function with adjustable parameters such as frequency (omega) and phase (phi). By employing the curve_fit function from scipy.optimize, we optimize the parameters to achieve the best fit. The resulting fitted curve is plotted alongside the original noisy data, providing insights into the effectiveness of the curve fitting process in approximating real-world data.


```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Define the function to fit
def f(t, omega, phi):
    return np.cos(omega * t + phi)

# Generate some noisy data
x = np.linspace(0, 3, 50)
y = f(x, 1.5, 1) + 1*np.random.normal(size=len(x))

# Fit the curve to the data using curve_fit
popt, pcov = curve_fit(f, x, y)

# Print the optimized parameters
print('Optimized Parameters:')
print('omega =', popt[0])
print('phi =', popt[1])

# Plot the original data and the fitted curve
y_pred = f(x, *popt)
plt.plot(x, y, 'b.')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Noisy Data')
plt.show()

plt.plot(x, y, 'b.', label='Noisy data')
plt.plot(x, y_pred, 'r-', label='Fitted curve')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Fitted Data')
plt.legend()
plt.show()
```

In this instance, x is a set of 250 values spanning from 0 to 3, and the y-values are calculated with f. (x, 1.5, 1). This suggests that we are considering the function f at omega=1.5 and phi=1.

If we alter the number of omega to 3, we get a distinct set of y-values, i.e., f (x, 3, 1). The magnitude of the cosine wave, in particular, should increase, resulting in a more oscillatory form.
